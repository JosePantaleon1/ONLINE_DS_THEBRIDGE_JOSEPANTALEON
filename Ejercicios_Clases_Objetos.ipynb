```python
class Tienda:
    """
    Clase que representa una tienda de electrodomésticos.

    Atributos de clase:
    - tipo: Especifica el tipo de productos que vende la tienda. En este caso, "Electrodomésticos".
    - abierta: Booleano que indica si la tienda está abierta o cerrada.

    Atributos de instancia:
    - nombre (str): Nombre de la tienda.
    - direccion (str): Dirección de la tienda.
    - num_empleados (int): Número de empleados que tiene la tienda.
    - ventas_3_meses (list): Lista de las ventas de los últimos tres meses.
    """
    
    tipo = "Electrodomésticos"
    abierta = True

    def __init__(self, nombre, direccion, numero_empleados, ventas_3_meses):
        """
        Inicializa la tienda con nombre, dirección, número de empleados y las ventas de los últimos tres meses.
        
        Args:
        - nombre (str): Nombre de la tienda.
        - direccion (str): Dirección de la tienda.
        - numero_empleados (int): Número de empleados.
        - ventas_3_meses (list): Lista con las ventas de los últimos tres meses.
        """
        self.nombre = nombre
        self.direccion = direccion
        self.num_empleados = numero_empleados
        self.ventas_3_meses = ventas_3_meses

    def ventas_3_meses_total(self):
        """
        Calcula el total de las ventas de los últimos tres meses.

        Devuelve:
        - (int): La suma de las ventas de los últimos tres meses.
        """
        total_ventas = sum(self.ventas_3_meses)
        print(f"Las ventas en los últimos 3 meses fueron: {total_ventas}")
        return total_ventas

    def ventas_por_empleado(self):
        """
        Calcula el promedio de ventas por empleado en los últimos tres meses.

        Devuelve:
        - (float): Promedio de ventas por empleado en los últimos tres meses.
        """
        total_ventas = sum(self.ventas_3_meses)
        promedio_por_empleado = total_ventas / self.num_empleados
        return promedio_por_empleado

    def nombre_direccion(self):
        """
        Imprime el nombre y la dirección de la tienda.
        """
        print(f"El nombre de la tienda es: {self.nombre}, y la dirección es: {self.direccion}")

    def ventas_ultimo_mes(self):
        """
        Imprime las ventas del último mes de los tres meses almacenados.
        """
        ventas_ultimo_mes = self.ventas_3_meses[-1]
        print(f"Las ventas del último mes fueron: {ventas_ultimo_mes}")
    
    def proyeccion_ventas(self, inversion_marketing):
        """
        Calcula y proyecta las ventas según la inversión en marketing.
        
        Si la inversión es menor a 1000, las ventas aumentan un 20%. Si es mayor o igual a 1000, 
        las ventas aumentan un 50%.

        Args:
        - inversion_marketing (int): La cantidad de dinero invertida en marketing.

        Devuelve:
        - (list): Las nuevas ventas proyectadas para los últimos 3 meses.
        """
        # Determinar el incremento en función de la inversión
        if inversion_marketing < 1000:
            incremento = 1.2
        else:
            incremento = 1.5
        
        # Actualizar las ventas con el incremento
        self.ventas_3_meses = [venta * incremento for venta in self.ventas_3_meses]
        return self.ventas_3_meses
```
