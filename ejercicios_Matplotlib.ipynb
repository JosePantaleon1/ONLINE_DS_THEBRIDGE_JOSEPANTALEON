```python
x = np.linspace(0, 5, 100)

# Definir las rectas
y1 = 3 * x  # Recta con pendiente positiva que pasa por (0,0)
y2 = -3 * x - 5  # Recta con pendiente negativa que pasa por (0,-5)

# Crear una figura con dos subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5), sharey=True)  # Compartir el eje Y

# Primera gr치fica: Recta con pendiente positiva
ax1.plot(x, y1, label='y = 3x', color='blue')
ax1.set_title('recta sube')  # T칤tulo de la primera gr치fica
ax1.set_xlabel('eje X')  # Etiqueta del eje X
ax1.set_ylabel('eje Y')  # Etiqueta del eje Y (solo en la primera gr치fica)
ax1.grid(True)
ax1.legend()

# Fijar l칤mites de los ejes
ax1.set_xlim(0, 5)
ax1.set_ylim(-15, 15)

# Segunda gr치fica: Recta con pendiente negativa
ax2.plot(x, y2, label='y = -3x - 5', color='red')
ax2.set_title('recta baja')  # T칤tulo de la segunda gr치fica
ax2.set_xlabel('eje X')  # Etiqueta del eje X
ax2.grid(True)
ax2.legend()

# Fijar l칤mites de los ejes
ax2.set_xlim(0, 5)

# Ajustar el espacio entre los subplots
plt.tight_layout()

# Mostrar la gr치fica
plt.show()
```

9. Bas치ndote en el gr치fico anterior, escribe encima de cada barra la altura de cada amigo.
Pista: usa un bucle que lea cada barra de barplot = plt.bar(x,y)
bar tiene los m칠todos get_height(), get_x(), get_width()
plt.text(x,y,valor, va='bottom')

```python
personas = ['Ana', 'Luis', 'Pedro', 'Sof칤a', 'Carmen']
alturas = [160, 180, 175, 190, 170]

# Crear el gr치fico de barras
barplot = plt.bar(personas, alturas, color='green')

# A침adir etiquetas y t칤tulo
plt.xlabel('Personas')
plt.ylabel('Altura (cm)')
plt.title('Altura de un grupo de amigos')

# A침adir la altura encima de cada barra
for bar in barplot:
    height = bar.get_height()  # Obtener la altura de la barra
    plt.text(bar.get_x() + bar.get_width() / 2, height, f'{int(height)}', 
             ha='center', va='bottom')  # Escribir el valor encima de la barra

# Mostrar la gr치fica
plt.show()
```

21. Vamos a pintar un histograma

```python

# Fijar una semilla para la reproducibilidad
np.random.seed(0)

# Generar una lista aleatoria de alturas (en cm) para 300 personas
# Suponiendo que la altura sigue una distribuci칩n normal
# Media de 170 cm y desviaci칩n est치ndar de 10 cm
num_personas = 300
altura_media = 170
desviacion_estandar = 10

alturas = np.random.normal(loc=altura_media, scale=desviacion_estandar, size=num_personas)

# Crear el histograma
plt.figure(figsize=(10, 6))
plt.hist(alturas, bins=30, color='blue', alpha=0.7, edgecolor='black')

# A침adir etiquetas y t칤tulo
plt.xlabel('Altura (cm)')
plt.ylabel('N칰mero de Personas')
plt.title('Histograma de Alturas de 300 Personas')

# Mostrar la cuadr칤cula
plt.grid(axis='y', alpha=0.75)

# Mostrar la gr치fica
plt.show()
```
